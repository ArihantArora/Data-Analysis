# Consider 2023 as the cutoff and calculate average usage in terms of Driven_km per year for each car. Get top 3 car brands and bottom 3 card brands as per usage. 

```python
car_data.info()
```

```python
car_data['Model_Year'] = car_data['Model_Year'].astype(int)
```

```python
def usage(dist):
    if dist['Model_Year'] == 2023:
        return dist['Distance']
    else:
        return dist['Distance']/(2023 - dist['Model_Year'])
```

```python
car_data['Usage'] = car_data.apply(usage,axis=1)
```

```python
car_data
```

```python
car_data.groupby('Car_brand')['Usage'].mean().sort_values(ascending=False).head(3)
```
